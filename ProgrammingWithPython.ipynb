```python
class CreateTable:
    
    '''
    A class to create SQL tables
    '''
    
    def __init__(self,name,dic,df,engine):
        '''
        A function that initializes attributes
        
        Parameters
        ----------
        name: table name
        dic: table columns
        df: table data
        engine
        
        '''
        self.name=name
        self.dic=dic
        self.df=df
        self.engine=engine
        
    def sqltable(self):
        '''
        A function that creates sql tables
        
        Parameters
        ---------
        self
        
        '''
        self.df.to_sql(
            self.name,
            self.engine,
            if_exists='replace',
            index=False,
            chunksize=500,
            dtype=self.dic    
        )
```

```python
class TableInfo(CreateTable):
    '''
    A class that gives all info on the Table created.
    It is a child class of 'CreateTable' and inherits all its attributes
    
    '''
    def __init__(self,name,dic,df,engine,shape,head,info):
        '''
        A function that initializes attributes
        
        Parameters
        ----------
        name: table name
        dic: table columns
        df: table data
        engine
        shape: table shape
        head: table initial rows
        info: table info
        
        '''
        super().__init__(name, dic, df, engine)
        self.shape = shape
        self.head = head
        self.info = info
        
shape_1 = df_train.shape
head_1 = df_train.head(10)
info_1 = df_train.info()
Table1info = TableInfo(table_name1, dic1, df_train, engine, shape_1, head_1, info_1)
Table1info.info
```
